```python
import matplotlib.pyplot as plt
import numpy as np

systems = ["Moderate-noise GRU", "Mouse"]
colors = ["#6a3d9a", "#1f77b4"]

hit_rates = [0.798442, 0.610860]
false_alarm_rates = [0.048744, 0.066667]
d_primes = [2.491848, 1.681003]
criteria = [0.410122, 0.560243]

x = np.arange(len(systems))
bar_width = 0.34

fig, axes = plt.subplots(
    1,
    2,
    figsize=(12, 4.8)
)

axes[0].bar(
    x - bar_width / 2,
    hit_rates,
    width=bar_width,
    color=colors,
    label="Hit rate"
)

axes[0].bar(
    x + bar_width / 2,
    false_alarm_rates,
    width=bar_width,
    color=["#b79ad1", "#8fc4e8"],
    label="False-alarm rate"
)

axes[0].set_xticks(x)
axes[0].set_xticklabels(systems)
axes[0].set_ylim(0, 1)
axes[0].set_ylabel("Response probability")
axes[0].set_title("Behavioral change detection")
axes[0].legend(frameon=False)

axes[1].bar(
    x - bar_width / 2,
    d_primes,
    width=bar_width,
    color=colors,
    label="Sensitivity ($d'$)"
)

axes[1].bar(
    x + bar_width / 2,
    criteria,
    width=bar_width,
    color=["#b79ad1", "#8fc4e8"],
    label="Criterion ($c$)"
)

axes[1].axhline(
    0,
    color="gray",
    linestyle="--",
    linewidth=1
)

axes[1].set_xticks(x)
axes[1].set_xticklabels(systems)
axes[1].set_ylabel("Signal-detection value")
axes[1].set_title("Sensitivity and response bias")
axes[1].legend(frameon=False)

fig.suptitle(
    "RNN and mouse: related change-detection behavior",
    y=1.03,
    fontsize=14
)

plt.tight_layout()

plt.savefig(
    OUTPUT_FIGURE_DIR / "final_rnn_mouse_behavior_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
